```python
rank = "甲"

match rank:
    case "甲":
        print("優")
    case "乙":
        print("良")
    case "丙":
        print("可")
    case "丁":
        print("不可")
    case _:
        print("None")
```

```python
obj = 1j

match obj:
    case int():
        print(abs(obj))
    case str():
        print(obj[0])
    case _:
        print(f"意図しないtypeです.\n{type(obj)}")
```
